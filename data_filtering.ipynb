```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data = pd.read_json('books.json')
```

```python
data
```

```python
data['price'] = data['price'].str.replace('.', '')
data['price'] = data['price'].str.replace(',', '.')
```

```python
data
```

```python
data.info()
```

```python
data['price'] = data['price'].astype(float)
data['pages'] = data['pages'].replace('', np.nan)
data['pages'] = data['pages'].astype('Int64')
data['weight'] = data['weight'].str.replace(' kg', '')
data['weight'] = data['weight'].replace('', np.nan)
data['weight'] = data['weight'].astype(float)
```

```python
data
```

```python
data['title'] = data['title'].replace('', np.nan)
data['author'] = data['author'].replace('', np.nan)
data['category'] = data['category'].replace('', np.nan)
data['binding'] = data['binding'].replace('', np.nan)
data['format'] = data['format'].str.replace('x', '*')
data['format'] = data['format'].str.replace('X', '*')
data['format'] = data['format'].str.replace(',', '.')
data['format'] = data['format'].str.replace('cm', '')
data['format'] = data['format'].str.replace('CM', '')
data['format'] = data['format'].str.replace('V', '')
data['format'] = data['format'].str.replace('..', '.')
data['format'] = data['format'].str.replace('**', '*')
data['format'] = data['format'].str.replace('?', '*')
data['format'] = data['format'].str.replace('55', '5.5')
data['format'] = data['format'].str.replace('69', '6.9')
data['format'] = data['format'].str.replace('96', '9.6')
data['format'] = data['format'].str.replace('140', '14.0')
data['format'] = data['format'].str.replace('145', '14.5')
data['format'] = data['format'].str.replace('147', '14.7')
data['format'] = data['format'].str.replace('148', '14.8')
data['format'] = data['format'].str.replace('200', '20.0')
data['format'] = data['format'].str.replace('210', '21.0')
data['format'] = data['format'].str.replace('250', '25.0')
data['format'] = data['format'].str.replace('297', '29.7')
data['format'] = data['format'].str.replace('320', '32.0')
data['format'] = data['format'].str.replace('420', '42.0')
data['format'] = data['format'].str.replace('13020', '13*20')
data['format'] = data['format'].str.replace('20525', '20')
data['format'] = data['format'].str.replace('20525', '20')
def remove_extra_multiply(value):
    if type(value) == float:
        return ''
    if type(value) == int:
        return ''
    if value.find('u') != -1: 
        return ''
    if value.find('tvrd') != -1: 
        return ''
    if value.find('B5') != -1: 
        return ''
    if value.find('A5') != -1: 
        return ''
    if len(value) < 3:
        return ''
    first_multiply_index = value.find('*')  # Find index of first '*'
    second_multiply_index = value.find('*', first_multiply_index + 1)  # Find index of second '*'
    if second_multiply_index != -1:  # If second '*' exists
        value = value[:second_multiply_index]  # Extract string till second '*'
    return value
    
def remove_year(value):
    if value == '':
        return value
    elif value == np.nan:
        return ''
    elif int(value) >= 2025:
        return ''
    elif int(value) < 1900:
        return ''
    else:
        return value

data['format'] = data['format'].apply(remove_extra_multiply)
data['format'] = data['format'].replace('', np.nan)
data['year'] = data['year'].apply(remove_year)
data['year'] = data['year'].replace('', np.nan)
data['year'] = data['year'].astype('Int64')
data['description'] = data['description'].replace('', np.nan)
data['publisher'] = data['publisher'].replace('', np.nan)
data['title'] = data['title'].str.strip()
```

```python
data
```

```python
data.dropna(inplace=True)
data.drop_duplicates(subset='title', inplace=True)
```

```python
data.info()
```

```python
filtered_df = data[data['price'] < 1500]

if len(filtered_df) >= 3000:
    sampled_df = filtered_df.sample(n=3000, random_state=1)
    
    sampled_indices = sampled_df.index
    
    data = data.drop(sampled_indices)
    
data
```

```python
data.to_json('books_filter.json', orient='records', force_ascii=False)
```
